```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, Input
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import yfinance as yf
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping

if tf.config.list_physical_devices('GPU'):
    print("GPU kullanılabilir.")
    tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)
else:
    print("GPU bulunamadı. CPU kullanılacak.")

symbols = ['KRDMA', 'KRDMB', 'KRDMD', 'EREGL', 'ISDMR', 'KCAER', 'YKSLN', 'CEMTS', 'CUSAN', 'ERBOS']
start_date = '2023-02-01'
end_date = '2024-01-01'

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
def prepare_data(data, feature='Close', time_steps=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data[feature].values.reshape(-1, 1))

    X, y = [], []
    for i in range(time_steps, len(data_scaled)):
        X.append(data_scaled[i-time_steps:i, 0])
        y.append(data_scaled[i, 0])  
    X, y = np.array(X), np.array(y)
    X = X.reshape(X.shape[0], X.shape[1], 1)
    return X, y, scaler

def build_lstm_model(input_shape):
    model = Sequential()
    model.add(Input(shape=input_shape))
    model.add(LSTM(units=200, return_sequences=True))
    model.add(Dropout(0.4))
    model.add(LSTM(units=200, return_sequences=True))
    model.add(Dropout(0.4))
    model.add(LSTM(units=200))
    model.add(Dropout(0.4))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_gru_model(input_shape):
    model = Sequential()
    model.add(Input(shape=input_shape))
    model.add(GRU(units=200, return_sequences=True))
    model.add(Dropout(0.4))
    model.add(GRU(units=200, return_sequences=True))
    model.add(Dropout(0.4))
    model.add(GRU(units=200))
    model.add(Dropout(0.4))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def format_time_axis(ax, num_points, time_labels):
    tick_positions = np.linspace(0, num_points - 1, len(time_labels))
    formatted_labels = time_labels[:len(tick_positions)]
    ax.set_xticks(tick_positions)
    ax.set_xticklabels(formatted_labels, rotation=45)

time_steps = 60
results = {}
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

for symbol in symbols:
    print(f"Processing {symbol}...")
    data = yf.download(symbol + '.IS', start=start_date, end=end_date, interval='1h')
    data.index = data.index.tz_convert('Europe/Istanbul')
    data = data.between_time('10:00', '18:00')
    if data.empty:
        print(f"{symbol} için yeterli veri yok.")
        continue

    X, y, scaler = prepare_data(data)
    split_index = int(len(X) * 0.8)
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    lstm_model = build_lstm_model((X_train.shape[1], 1))
    lstm_model.fit(X_train, y_train, epochs=40, batch_size=32, verbose=1, validation_data=(X_test, y_test), callbacks=[early_stopping])
    lstm_predictions = lstm_model.predict(X_test)
    lstm_predictions = scaler.inverse_transform(lstm_predictions)
    real_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
    gru_model = build_gru_model((X_train.shape[1], 1))
    gru_model.fit(X_train, y_train, epochs=40, batch_size=32, verbose=1, validation_data=(X_test, y_test), callbacks=[early_stopping])
    gru_predictions = gru_model.predict(X_test)
    gru_predictions = scaler.inverse_transform(gru_predictions)

    mse_lstm = mean_squared_error(real_prices, lstm_predictions)
    rmse_lstm = np.sqrt(mse_lstm)
    mae_lstm = mean_absolute_error(real_prices, lstm_predictions)
    mape_lstm = mean_absolute_percentage_error(real_prices, lstm_predictions)
    r2_lstm = r2_score(real_prices, lstm_predictions)
    mse_gru = mean_squared_error(real_prices, gru_predictions)
    rmse_gru = np.sqrt(mse_gru)
    mae_gru = mean_absolute_error(real_prices, gru_predictions)
    mape_gru = mean_absolute_percentage_error(real_prices, gru_predictions)
    r2_gru = r2_score(real_prices, gru_predictions)

    print(f"{symbol} LSTM Model Metrics:")
    print(f"MSE: {mse_lstm:.4f}")
    print(f"RMSE: {rmse_lstm:.4f}")
    print(f"MAE: {mae_lstm:.4f}")
    print(f"MAPE: {mape_lstm:.4f}")
    print(f"R2: {r2_lstm:.4f}")
    
    print(f"{symbol} GRU Model Metrics:")
    print(f"MSE: {mse_gru:.4f}")
    print(f"RMSE: {rmse_gru:.4f}")
    print(f"MAE: {mae_gru:.4f}")
    print(f"MAPE: {mape_gru:.4f}")
    print(f"R2: {r2_gru:.4f}")

    fig, ax = plt.subplots(figsize=(12, 6))
    time_labels = [f"{hour}:00" for hour in range(10, 19)]
    

    ax.plot(range(len(real_prices[-8:])), real_prices[-8:], label='Gerçek Fiyatlar - Son 1 Gün', color='blue')
    ax.plot(range(len(lstm_predictions[-8:])), lstm_predictions[-8:], label='LSTM Tahminleri - Son 1 Gün', color='red')
    ax.plot(range(len(gru_predictions[-8:])), gru_predictions[-8:], label='GRU Tahminleri - Son 1 Gün', color='green')
    ax.plot(range(len(real_prices[-40:])), real_prices[-40:], label='Gerçek Fiyatlar - Son 5 Gün', color='blue', linestyle='--')
    ax.plot(range(len(lstm_predictions[-40:])), lstm_predictions[-40:], label='LSTM Tahminleri - Son 5 Gün', color='red', linestyle='--')
    ax.plot(range(len(gru_predictions[-40:])), gru_predictions[-40:], label='GRU Tahminleri - Son 5 Gün', color='green', linestyle='--')
    num_points = len(real_prices[-160:])
    ax.plot(range(num_points), real_prices[-160:], label='Gerçek Fiyatlar - Son 20 Gün', color='blue', linestyle=':')
    ax.plot(range(num_points), lstm_predictions[-160:], label='LSTM Tahminleri - Son 20 Gün', color='red', linestyle=':')
    ax.plot(range(num_points), gru_predictions[-160:], label='GRU Tahminleri - Son 20 Gün', color='green', linestyle=':')
    ax.set_title(f"{symbol} Tahminleri ve Gerçek Fiyatlar")
    ax.set_xlabel('Zaman (Saat)')
    ax.set_ylabel('Fiyat', color='tab:blue')
    format_time_axis(ax, num_points, time_labels)
    ax.legend()
    plt.tight_layout()
    plt.show()
```
