```python
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, f1_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout
import yfinance as yf

if tf.config.list_physical_devices('GPU'):
    print("GPU kullanılabilir.")
    tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)
else:
    print("GPU bulunamadı. CPU kullanılacak.")

symbols = ['KRDMA', 'KRDMB', 'KRDMD', 'EREGL', 'ISDMR', 'KCAER', 'YKSLN', 'CEMTS', 'CUSAN', 'ERBOS']
start_date = '2015-01-01'
end_date = '2024-01-01'
company_data = {}
for symbol in symbols:
    data = yf.download(symbol + '.IS', start=start_date, end=end_date)
    company_data[symbol] = data
def prepare_data(data, feature='Close', time_steps=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data[feature].values.reshape(-1, 1))

    X, y = [], []
    for i in range(time_steps, len(data_scaled)):
        X.append(data_scaled[i-time_steps:i, 0])
        y.append(data_scaled[i, 0])

    X, y = np.array(X), np.array(y)
    X = X.reshape(X.shape[0], X.shape[1], 1)
     
    return X, y, scaler

def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_gru_model(input_shape):
    model = Sequential()
    model.add(GRU(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(GRU(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(GRU(units=50))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

time_steps = 60
test_size = 0.2
results = {}
     
for symbol, data in company_data.items():
    print(f"Processing {symbol} with LSTM...")
    X, y, scaler = prepare_data(data)
    split_index = int(len(X) * (1 - test_size))
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]
    lstm_model = build_lstm_model((X_train.shape[1], 1))
    lstm_model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)
    lstm_predictions = lstm_model.predict(X_test)
    lstm_predictions = scaler.inverse_transform(lstm_predictions)
    real_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

    mse = mean_squared_error(real_prices, lstm_predictions)
    mae = mean_absolute_error(real_prices, lstm_predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(real_prices, lstm_predictions)

    threshold = np.mean(real_prices)
    binary_real = (real_prices > threshold).astype(int)
    binary_pred = (lstm_predictions > threshold).astype(int)
    accuracy = accuracy_score(binary_real, binary_pred) * 100
    f1 = f1_score(binary_real, binary_pred)
    print(f"{symbol} LSTM - MSE: {mse:.2f}, MAE: {mae:.2f}, RMSE: {rmse:.2f}, R^2: {r2:.2f}, Accuracy: {accuracy:.2f}%, F1 Score: {f1:.2f}")
    results[symbol] = {'real_prices': real_prices, 'lstm_predictions': lstm_predictions}

    plt.figure(figsize=(10, 5))
    plt.plot(real_prices, color='blue', label='Real Prices')
    plt.plot(lstm_predictions, color='red', label='LSTM Predicted Prices')
    plt.title(f'{symbol} Stock Price Prediction with LSTM')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
    print(f"Processing {symbol} with GRU...")

    gru_model = build_gru_model((X_train.shape[1], 1))
    gru_model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)
    gru_predictions = gru_model.predict(X_test)
    gru_predictions = scaler.inverse_transform(gru_predictions)

    mse = mean_squared_error(real_prices, gru_predictions)
    mae = mean_absolute_error(real_prices, gru_predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(real_prices, gru_predictions)

    binary_pred = (gru_predictions > threshold).astype(int)
    accuracy = accuracy_score(binary_real, binary_pred) * 100
    f1 = f1_score(binary_real, binary_pred)
    print(f"{symbol} GRU - MSE: {mse:.2f}, MAE: {mae:.2f}, RMSE: {rmse:.2f}, R^2: {r2:.2f}, Accuracy: {accuracy:.2f}%, F1 Score: {f1:.2f}")
    results[symbol]['gru_predictions'] = gru_predictions

    plt.figure(figsize=(10, 5))
    plt.plot(real_prices, color='blue', label='Real Prices')
    plt.plot(gru_predictions, color='green', label='GRU Predicted Prices')
    plt.title(f'{symbol} Stock Price Prediction with GRU')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

print("Processing complete for all companies.")
```
